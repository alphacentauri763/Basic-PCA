```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng = np.random.RandomState(0)
```

```python
W = rng.randn(1000, 2)
```

```python
plt.scatter(W[:,0], W[:, 1])
plt.axis('equal')
plt.grid('on')
```

```python
A = rng.rand(2, 2)
```

```python
X = np.dot(W, A) 
```

```python
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
plt.grid('on')
```

```python
np.var(X[:,0])
```

```python
np.var(X[:,1])
```

```python
C = np.cov(X.T)
C
```

```python
e, V = np.linalg.eig(C)
```

```python
e
```

```python
V
```

```python
V[:,0]
```

```python
V[:,1]
```

```python
u = V[:,1]
```

```python
u.shape
```

```python
Z = np.dot(X, u)
```

```python
Z.shape
```

```python
np.var(Z)
```

```python
np.dot(u,u)
```

```python
Y = np.outer(Z,u)
```

```python
Y.shape
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(Y[:,0], Y[:,1])
plt.axis('equal')
plt.grid('on')
```
